## Addition of two numbers

```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph # type: ignore
```

```python
class AgentState(TypedDict):
    n1 : int
    n2 : int
    sum: int

def calculate_node(state: AgentState) -> AgentState:
    """ Addition of two numbers """
    state['sum'] = state['n1'] + state['n2'] 
    state['sum'] = "Addition of two numbers: " + str(state['sum'])
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node('Calculate', calculate_node)

graph.set_entry_point('Calculate')

graph.set_finish_point('Calculate')

app = graph.compile()
```

```python
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({'n1': 5, 'n2': 10})

result["sum"]
```
