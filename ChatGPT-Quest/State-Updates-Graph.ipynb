```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    number: int
    result: int
```

```python
def operation(state: AgentState) -> AgentState:
    """This node performs a simple operation on the input number."""

    state['result'] = state['number'] * 2
    state['result'] += 5
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("operation", operation)
graph.set_entry_point("operation")
graph.set_finish_point("operation")

app = graph.compile()
```

```python
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
answer = app.invoke({"number" : 30, "result": 0})
print(answer['result'])
```
